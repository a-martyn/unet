# Test dataloader

```python
train_pth = 'data/membrane/train'
test_pth = 'data/membrane/test'
```

```python
def show_sample(generator):
    batch = next(generator)
    x = batch[0][0]
    y = batch[1][0]
    
    size = (5, 5)
    plt.figure(figsize=size)
    plt.imshow(x[:, :, 0], cmap='gray')
    plt.show()
    plt.figure(figsize=size)
    plt.imshow(y[:, :, 0], cmap='gray')
    plt.show();
    return
```

```python
# Data augmentation
data_gen_args = dict(
    rotation_range=0.2,
    width_shift_range=0.05,
    height_shift_range=0.05,
    shear_range=0.05,
    zoom_range=0.05,
    horizontal_flip=True,
    fill_mode='reflect'
)
```

## Train generator with transforms

```python
gen_train = generator(train_pth, transforms=data_gen_args)
```

```python
show_sample(gen_train)
```

## Test generator without transforms

Note some preprocessing was required to force test targets to render correctly. Some information in test masks may have been lost. See `preprocessing.py` for details.

```python
gen_test = generator(test_pth)
```

```python
show_sample(gen_test)
```
